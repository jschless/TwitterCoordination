```python
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
TWITTER_DATA_DIR = '/home/joe/Dropbox/MIT/Thesis/grid_files/campaigns'
flag = 0
for f_name in os.listdir(TWITTER_DATA_DIR):
    if f_name.endswith('results.pkl'):
        with open(os.path.join(TWITTER_DATA_DIR, f_name), 'rb') as f:
            res = pickle.load(f)
            N_NEIGHBORS = res['n_neighbors']
            templates = res['template_n_template_neighbors']
            normals = res['normal_n_template_neighbors']
            if len(normals) < 3000:
                continue
            global_pct_temp = res['global_pct_temp']
            temp_pcts = np.array(templates)/N_NEIGHBORS
            norm_pcts = np.array(normals)/N_NEIGHBORS
            
            temp_pcts /= global_pct_temp
            norm_pcts /= global_pct_temp
            ecdf_temp = ECDF(temp_pcts)
            ecdf_norm = ECDF(norm_pcts)
  
            print(N_NEIGHBORS)
#             try: 
#                 if f_name == 'merapmmeraabhimaan_neighbors_results.pkl' and ecdf_temp.x[10] < 1.4:
#                     #print(ecdf_temp.x, ecdf_temp.y)
#                     print(N_NEIGHBORS)
#                     print(global_pct_temp)
#                     print(f_name)
                    
#                 else:
#                     continue
#             except:
#                 continue
#                 print()
            if flag:
                label_1, label_2 = '', ''
            else:
                label_1, label_2 = 'Template', 'Normal'
                
            max_neighborhood_size = 110
            size = len(ecdf_temp.x) + len(ecdf_norm.x)
            #print(size/100/max_neighborhood_size)
            print(N_NEIGHBORS/max_neighborhood_size)
            plt.plot(ecdf_temp.x, ecdf_temp.y, label=label_1, color='r', 
                     lw=(N_NEIGHBORS/max_neighborhood_size))
            plt.plot(ecdf_norm.x, ecdf_norm.y, label=label_2, color='b', alpha=.6)
            flag = 1
plt.xlim(0,5)
plt.xticks([0,.5,1,2,5])
plt.yticks([0,.25,.5,.75,1])
plt.legend()
plt.title('ECDFs of Template Penetration Rate by Tweet Type')
plt.xlabel('Normalized Template Penetration Rate')
plt.savefig('temp.jpeg')
```
