# Panel Data
Ok, so I want to include the following options:
Treatment time:
    
    - earliest
    - best 
    - donut hole uncertainty 
    
Trending location cutoffs:
    - 10
    - 25
    - 50 (None)

```python
cutoff_choices = ['best', 'earliest']
trending_cutoffs = [10, 25, 50]
time_bins = ['1Min', '5Min', '10Min']
include_missings = [True, False]
user_exposes_himselfs = [True, False]

options = list(itertools.product(cutoff_choices, trending_cutoffs, time_bins, 
                                 include_missings, user_exposes_himselfs))
print('# options', len(options))
```

```python
df_dict = {}
for opt in tqdm(list(itertools.product(cutoff_choices, trending_cutoffs, time_bins, 
                                 include_missings, user_exposes_himselfs))):
    cutoff_choice, trending_thresh, time_bin, include_missing, user_exposes_himself = opt
    dfs = []
    for hashtag, tweet_dict in campaigns.items():
        if hashtag not in preprocessing.bandwagon_hashtags and hashtag not in preprocessing.never_trended:
            temp, _ = trending.build_df(hashtag, tweet_dict, exposures, normalize_time=True, time_bin=time_bin,
                                        cutoff_choice=cutoff_choice, include_missing=include_missing, 
                                        user_exposes_himself=user_exposes_himself, 
                                        trending_loc_cutoff=trending_thresh)
            
            if temp is not None:
                temp = temp.assign(threshold=(temp.time > 0).astype(int))
                temp['hashtag'] = hashtag
                dfs.append(temp)
        
    df_dict[opt] = pd.concat(dfs)
```

## Experimentation

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.discrete import count_model as cm
from statsmodels.discrete import discrete_model as dm
import statsmodels.api as sm

def log(x):
    return np.log(x+1)
    if x == 0:
        return 0
    else:
        return np.log(x)

def prepare_data(df, logged=False):
    df['timeXthreshold'] = df['time'] * df['threshold']
    if logged:
        for col in ['regular', 'nonzero_exposure_regular']:
            df['ln_'+ col] = df[col].apply(log)

    endog = df['ln_regular'] if logged else df['regular']

    last_var = 'nonzero_exposure_regular'
    if logged:
        last_var = 'ln_' + last_var
        
    exog = df[['time', 'threshold', 'timeXthreshold', last_var]]
    exog['const'] = 1
    exog = pd.concat((exog, pd.get_dummies(df['hashtag'])), axis=1).fillna(0)
    return endog, exog
```

```python
result_dict = {}

for opts, panel_df in tqdm(df_dict.items()):
    for model_type in [cm.ZeroInflatedPoisson, dm.Poisson, OLS]:
        cutoff_choice = opts[0]
        trending_thresh = opts[1]

        for end_time in [60, 120, 180, 240]:        
            query = panel_df.query(f'time >= -{end_time} and time <= {end_time}')
            query.hashtag = pd.Categorical(query.hashtag).codes
            try:
                is_ols = model_type.__name__ == "OLS"
                mod = model_type(*prepare_data(query, logged=is_ols)).fit(cov_type='HC1', disp=False)
                result_dict[(model_type.__name__, *opts, end_time)] = mod
            except Exception as e:
                print('model failed on settings', model_type.__name__, opts, e)
            # donut hole!
            if cutoff_choice == 'earliest' and end_time != 60:
                query  = query.query('time <= 0 or time > 60')
                try:
                    mod = model_type(*prepare_data(query, logged=is_ols)).fit(cov_type='HC1', disp=False)
                    result_dict[(model_type.__name__, 'Donut Hole', *opts[1:], end_time)] = mod
                except Exception as e:
                    print('model failed on settings', model_type.__name__, opts, e)
```

```python
from statsmodels.iolib.summary2 import summary_col


res_df = summary_col(list(result_dict.values()), 
                      model_names=['-'.join([str(t) for t in x]) for x in result_dict.keys()],
                      stars=True).tables[0]

ind = res_df.index.get_loc('threshold')

res_df = (res_df.iloc[ind:ind+2, :].T
                .rename(columns={'threshold': 'coef', '': 'stdev'})
         )

for i, col in enumerate(['Model_Type', 'Cutoff_Choice', 'Trending_Cutoff', 
                         'Time_Bin', 'Include_Missing', 
                         'Include_Multiple_Exposures', 'Window_Length']):
    res_df[col] = [x.split('-')[i] for x in res_df.index]
    
res_df['stdev'] = res_df.stdev.apply(lambda x: float(x[1:-1]))
res_df['n_stars'] = res_df.coef.apply(lambda x: x.count('*'))
res_df['coef'] = res_df.coef.apply(lambda x: float(x.replace('*', '')))
res_df['Include_Missing'] = res_df.Include_Missing.apply(eval)
res_df['Include_Multiple_Exposures'] = res_df.Include_Multiple_Exposures.apply(eval)

res_df = (res_df.astype({'Trending_Cutoff': int, 'Window_Length': int})
                .sort_values('coef')
         )

res_df = res_df.reset_index()
```

```python
import matplotlib.patches as patches
import math
import matplotlib.ticker as ticker

def spec_curve(res_df, title='Trending Effect Specification Curve'):
    fig = plt.figure(constrained_layout=False, figsize=(22, 12))
    print('# Trials', len(res_df))
    spec = fig.add_gridspec(ncols=1, nrows=count,
                            height_ratios=[20] + [1]*(count-1), wspace=0.05)
    axarr = []
    for row in range(count):
        axarr.append(fig.add_subplot(spec[row, 0]))


    star_map = {0: 'red', 1: 'red',
                2: 'b', 3: 'b'}

    res_df['color'] = res_df.n_stars.apply(lambda x: star_map.get(x))
    
    median = np.median(res_df['coef'])
    print('Median:', median)
    axarr[0].axhline(y=median,
                     color='black',
                     lw=0.8,
                     alpha=1,
                     label='Median coefficient',
                     dashes=[12, 5])

    for color in res_df['color'].unique():
        slice_df_r = res_df.loc[res_df['color'] == color]
        markers, caps, bars = axarr[0].errorbar(slice_df_r.index,
                                                slice_df_r.coef,
                                                yerr=slice_df_r.stdev,
                                                ls='none', color=color,
                                                alpha=0.8, zorder=1,
                                                elinewidth=2,
                                                marker='o')
        [bar.set_alpha(0.4) for bar in bars]
        [cap.set_alpha(0.4) for cap in caps]

    axarr[0].legend(frameon=True, loc='upper left',
                    ncol=1, handlelength=2)
    axarr[0].set_ylabel('Coefficient')
    axarr[0].set_title(title)
    max_height = (res_df.coef + res_df.stdev).max()
    min_height = (res_df.coef - res_df.stdev).min()
    ylims = (min_height/1.2, 1.2*max_height)
    print(ylims)
    axarr[0].set_ylim(round(ylims[0] - .09, 1),
                      round(ylims[1] + .09, 1)) # floor and ceiling

    # Now do the blocks - each group get its own array
    wid = 0.6
    hei = wid/3
    color_dict = {True: '#9B9B9B', False: '#F2F2F2'}
    # if(len(res_df) > 160):
    #     wid = 0.01
    #     color_dict = {True: 'k', False: '#FFFFFF'}

    for ax_num, ax in enumerate(axarr[1:]):
        col = mapping[ax_num]
        to_shade = res_df[col].apply(lambda x: x == options[ax_num])
        for j, val in enumerate(to_shade):
            color = color_dict[val]
            sq = patches.Rectangle(
                (j-wid/2, -hei/2), wid, hei, fill=True, color=color)
            ax.add_patch(sq)
        ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
        ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
        ax.set_yticks([0]) #hei]) 
        formatted_label = col.replace('_', ' ')
        formatted_dict = {'earliest': 'Earliest', 'best': 'Best', 'Donut Hole': 'Donut Hole',
                      '5Min': '5 Min', '10Min': '10 Min',
                      10: 'Top 10', 25: 'Top 25', 50: 'Top 50',
                      60: '60 Min', 120: '120 Min', 180: '180 Min', 240: '240 Min', 
                    'Poisson': 'Poisson', 'ZeroInflatedPoisson': 'ZIP', 'OLS': 'Log-linear'}
        if not isinstance(options[ax_num], bool): # don't include booleans
            formatted_label += '=' + formatted_dict[options[ax_num]]        
        ax.set_yticklabels([formatted_label])
        ax.set_xticklabels([])
        ax.set_ylim(-hei, 1-hei*4)
        ax.set_xlim(-wid, len(res_df))
        for place in ['right', 'top', 'bottom']:
            ax.spines[place].set_visible(False)
    for ax in axarr:
        ax.set_xticks([], minor=True)
        ax.set_xticks([])
        ax.set_xlim(-wid, len(res_df))
    plt.tight_layout()
```

# Logged Outcome

```python
ln_result_dict = {}
formula_string = "ln_regular ~ time*threshold + ln_nonzero_exposure_regular"

for opts, panel_df in tqdm(df_dict.items()):
    cutoff_choice = opts[0]
    trending_thresh = opts[1]
    
    for end_time in [60, 120, 180, 240]:        
        query = panel_df.query(f'time >= -{end_time} and time <= {end_time}')
        query.hashtag = pd.Categorical(query.hashtag).codes
        query['ln_regular'] = np.log(query.regular + 1)
        query['ln_nonzero_exposure_regular'] = np.log(query.nonzero_exposure_regular + 1)

        try:
            mod = smf.mixedlm(formula_string, data=query, groups=query['hashtag']).fit()
            ln_result_dict[(*opts, end_time)] = mod
        except Exception:
            print('model failed on settings', opts)
        # donut hole!
        if cutoff_choice == 'earliest' and end_time != 60:
            query  = query.query('time <= 0 or time > 60')
            query['ln_regular'] = np.log(query.regular + 1)
            query['ln_nonzero_exposure_regular'] = np.log(query.nonzero_exposure_regular + 1)
            try:
                mod = smf.mixedlm(formula_string, data=query, groups=query['hashtag']).fit()
                ln_result_dict[('Donut Hole', *opts[1:], end_time)] = mod
            except Exception:
                print('model failed on settings', opts)
```

```python
from statsmodels.iolib.summary2 import summary_col


res_df = summary_col(list(ln_result_dict.values()), 
                      model_names=['-'.join([str(t) for t in x]) for x in ln_result_dict.keys()],
                      stars=True).tables[0]

ind = res_df.index.get_loc('threshold')

res_df = (res_df.iloc[ind:ind+2, :].T
                .rename(columns={'threshold': 'coef', '': 'stdev'})
         )

for i, col in enumerate(['Cutoff_Choice', 'Trending_Cutoff', 
                         'Time_Bin', 'Include_Missing', 
                         'Include_Multiple_Exposures', 'Window_Length']):
    res_df[col] = [x.split('-')[i] for x in res_df.index]
    
res_df['stdev'] = res_df.stdev.apply(lambda x: float(x[1:-1]))
res_df['n_stars'] = res_df.coef.apply(lambda x: x.count('*'))
res_df['coef'] = res_df.coef.apply(lambda x: float(x.replace('*', '')))
res_df['Include_Missing'] = res_df.Include_Missing.apply(eval)
res_df['Include_Multiple_Exposures'] = res_df.Include_Multiple_Exposures.apply(eval)

res_df = (res_df.astype({'Trending_Cutoff': int, 'Window_Length': int})
                .sort_values('coef')
         )

res_df = res_df.reset_index()
res_df.head()

ln_res_df = res_df.copy()
```
